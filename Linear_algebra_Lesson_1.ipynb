5. Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;

б) все векторы, являющиеся линейными комбинациями данных векторов {u1,u2,…,un}.

Решение:
а) = нет, так как не соотвествует условию:
(a,0,c)+(0,b, d)=(a,b,c+d)

б) - да

Любой вектор множества можно представить как

$\sum_{t=1}^{n}\lambda_{t}u_{t}$

Сумма двух векторов:

$\sum_{t=1}^{n}\alpha_{t}u_{t} + \sum_{t=1}^{n}\beta_{t}u_{t} = \sum_{t=1}^{n}(\alpha_{t} + \beta_{t})u_{t} =\sum_{t=1}^{n}\lambda_{t}u_{t}$

Произведение вектора на число:

$\alpha \cdot \sum_{t=1}^{n}\lambda_{t}u_{t} = \sum_{t=1}^{n}\alpha \cdot \lambda_{t}u_{t} = \sum_{t=1}^{n}\beta_{t}u_{t}$

Следовательно, данные вектора принадлежат указанному множеству, поэтому это множество векторов является линейным подпространством.


Методичка №2

__1.__ Найти скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$


```python
import numpy as np
from numpy.linalg import norm
```

```python
x = np.array([0, -3, 6])
y = np.array([-4, 7, 9])
print("a * b =", np.dot(x, y))
```

```python
x = np.array([7, -4, 0, 1])
y = np.array([-3, 1, 11, 2])
print("a * b =", np.dot(x, y))
```


__2.__ Найти нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.



```python
a = np.array([4, 2, 4])
b = np.array([12, 3, 4])

print(norm(a))
print(norm(b))

cos_phi = np.dot(a, b) / norm(a) / norm(b)
print(f'Косинус угла векторов векторами {cos_phi:.2f}')
print(f'Угол {np.arccos(cos_phi):.2f}')
```

__4.__ Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

```python
# а)
a = np.array([1, 0, 0])
b = np.array([0, 0, 1])
print('Ортонормированный базис', np.dot(a, b))
```

```python
# б)
c = np.array([(1/(2**0.5)), (-1/(2**0.5)), 0])
d = np.array([(1/(2**0.5)), (1/(2**0.5)), 0])
b = np.array([0, 0, 1])
print('Ортонормированный базис')
print(np.dot(c, d))
print(np.dot(c, b))
print(np.dot(b, d))
print(f'{norm(c):.1f}, {norm(d):.1f}, {norm(b):.1f}')
```

```python
# в)
e = np.array([0.5, -0.5, 0])
f = np.array([0, 0.5, 0.5])
b = np.array([0, 0, 1])
print('Не ортогональный, не ортонормированный базис')
print(np.dot(e, f))
print(np.dot(e, b))
print(np.dot(b, f))
print(norm(e), norm(f), norm(b))
```
