2. Определитель матрицы $A$ равен $4$. Найти:

а) $det(A^{2})$;

б) $det(A^{T})$;

в) $det(2A)$.

Решение:

а) $det(A^{2})$ = $det(A) \times det(A)$ = 16

б) $det(A^{T})$ = $det(A)$ = 4

в) $det(2A)$ = $2^{rank(A)} \times det(A)$ = $2^{rank(A)} \times 4$ = $2^{rank(A) + 2}$

3. Доказать, что матрица

 $
 \begin{pmatrix}
-2 & 7 & -3\\ 
4 & -14 & 6\\ 
-3 & 7 & 13
\end{pmatrix}
 $
 
вырожденная.

```python
rank = np.linalg.matrix_rank
inv = np.linalg.inv

A = np.matrix('-2, 7, -3; 4, -14, 6; -3, 7, 13')
print(f'Ранг матрицы: {rank(A)}')
try:
    print(f'Обратная матрица: {inv(A)}')
except:
    print('Нет обратной матрицы')
```

Ранг матрицы меньше числа ее строк, и у нее отсутствует обратная матрица => Матрица 

$
 $
 \begin{pmatrix}
-2 & 7 & -3\\ 
4 & -14 & 6\\ 
-3 & 7 & 13
\end{pmatrix}
 $
$

вырождена
