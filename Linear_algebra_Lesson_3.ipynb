
__1.__ Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей


$$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$$

 

```python
import numpy as np
```

```python
def is_eig(A):
    w, v = np.linalg.eig(A)
    print(f'Матрица:\n{A}')
    print(f'Собственные векторы:\n{v}')
    print(f'Собственные значения:\n{w}')
```

```python
matrix_a = np.array([[-1, -6], [2, 6]])
is_eig(matrix_a)
```

__3.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.


$\begin{cases}
1+1=1\cdot\lambda \\ 
-1+3 = 1\cdot \lambda
\end{cases}
\Rightarrow
\begin{cases}
\lambda=2\\ 
\lambda=2
\end{cases}.$
=> Данный вектор является собственным вектором линейного оператора
